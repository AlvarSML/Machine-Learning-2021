<h1>Trabalho dataset titanic</h1>


```python
#importar el dataset
import pandas as pd
import seaborn as sns
```

```python
titanic=pd.read_csv('titanic-dataset.csv',index_col='PassengerId')
titanic.info()
```

```python
sns.jointplot(data=titanic,
             x="Fare",
             y="Age",
             hue="Pclass")
```

```python
sns.catplot(data=titanic,
            kind='violin',
            x='Survived',
            y='Pclass',
            hue='Sex',
            split='true')
```

```python
sns.pairplot(data=titanic,hue='Pclass')
```
